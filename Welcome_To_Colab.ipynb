```
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```
team = ['4343','4347','4344','5845','6904','4342','4788','4341','6903','4346','4787','4345','5843','4340','5143']
```

```
def get_average(teamcodes,batorbowl):
  flag = False
  for i in team:
    url = 'https://www.espncricinfo.com/records/trophy/averages-'+str(batorbowl)+'/indian-premier-league-117?team='+str(i)
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html')
    table = soup.find_all("table")[0]

    if flag == False:
      table_header = table.find('thead')
      table_header_rows = table_header.find('tr')
      stats_header = []
      for row in table_header_rows:
        stats_header.append(row.text)
      main_df = pd.DataFrame(columns = stats_header)
      flag = True

    table_body = table.find('tbody')
    table_body_rows = table_body.find_all('tr')
    stats_data = []

    for row in table_body_rows:
      row_data = row.find_all('td')
      ind_rows = [data.text.strip() for data in row_data]
      stats_data.append(ind_rows)

    df = pd.DataFrame(stats_data, columns=stats_header)

    header = soup.find_all('h1')[0]
    team_name = header.text.strip().split('- ')[1]
    df['team']=team_name

    main_df = pd.concat([main_df,df])
    return main_df
```

```
batting_stats = get_average(team,'batting')
```

```
bowling_stats = get_average(team,'bowling')
```
